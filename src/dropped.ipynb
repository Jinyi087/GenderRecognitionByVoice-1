```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

sns.set(font_scale=1.5)

data = pd.read_csv('data/cleaned/voice_data.csv')
X_train = pd.read_csv('data/cleaned/X_dropped_train.csv')
X_test = pd.read_csv('data/cleaned/X_dropped_test.csv')
y_train = pd.read_csv('data/cleaned/y_dropped_train.csv')
y_test = pd.read_csv('data/cleaned/y_dropped_test.csv')

print(X_train.shape)
```

```python
from sklearn.svm import SVC

svc = SVC().fit(X_train, y_train.values.ravel())
svc.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier().fit(X_train, y_train.values.ravel())
tree.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier().fit(X_train, y_train.values.ravel())
forest.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

print("original shape:   ", X_train.shape)
print("transformed shape:", X_train_pca.shape)
```

```python
plt.figure(figsize=(20, 20))
sns.scatterplot(x=X_train_pca[:, 0], y=X_train_pca[:, 1],  hue=y_train.values.ravel())
```
